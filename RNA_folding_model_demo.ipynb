A simulation starts by generating a random RNA sequence and defining its minimum free-energy secondary structure as the **reference structure**. 

The fitness of RNA sequence $i$ is determined using the function:

$$w_{i} = \left\{\begin{array}{ll}
        1 & \quad  \mathrm{if}  \quad \beta_i>\alpha \quad \mathrm{and} \quad  \delta_i \leqslant \alpha\\
        0 & \quad  \mathrm{otherwise}   \\
        \end{array}\right.$$
        
where $\beta_i$ is the number of base pairs in the secondary structure of sequence $i$, $\delta_{i}$ is the base-pair distance between the structure of sequence $i$ and the reference structure, and $\alpha$ is an arbitrary threshold.

To illustrate our code, we pick the first sequence in [Ref_seqs.txt](Ref_seqs.txt) (containing the $10^3$ reference sequences used in the simulations with $\alpha=12$).

```python
ref_seq = 'UAGACUAUAAAGUACCGGGAUCUAUUGAAUGUUGGUACCCCUUUCGAGUGAUUGCGCCCCAGUCCGUGAACUAGGAUGCAUAUUUACCUAUUCAGAGGAU'
```

We define an `Evolution` object and set $\alpha = 12$.

```python
evo = Evolution(ref_seq, alpha=12)
```

We run a simulation after setting the following parameters:

* `divergence`: the number of substitutions during evolution 

* `burnin`: the length of the burn-in period

```python
evo.evolve(divergence=40, burnin=200, verbose=True)
```

The `stats` attribute is a dictionary containing the simulation data that can be saved directly to file.

```python
results = evo.stats
```

The following shows the accumulation of simple DMIs in this simulation.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(np.arange(1, 41, 1), results['SDMI_count'], clip_on=False)
plt.ylabel('Simple DMIs')
plt.xlabel('Substitution');
```

## Model fitting

We fit a linear model:
$$I_k = (k - 1)b \quad ,$$ 
and the snowball model:
$$I_k = \frac{k(k-1)p}{2} \quad ,$$
where 
$I_k$ is the number of simple DMIs,
$k$ is the number of substitutions, and
$p$ and $b$ are parameters determining the rate of accumulation of simple DMIs.

We fit the models using R.

```python
%load_ext rpy2.ipython
```

```python
x =  np.arange(1, 41, 1)
y = results['SDMI_count']
```

```python
%%R -i x -i y -o b
lin = nls(y ~ b * (x - 1), start=list(b=.1))
b = summary(lin)$coef[1, 1] 
summary(lin)
```

```python
%%R -i x -i y -o p
orr = nls(y ~ p * x * (x - 1)/2, start=list(p=0.1))
p = summary(orr)$coef[1, 1]
summary(orr)
```

```python
plt.plot(x, y, clip_on=False)
plt.plot(x, [p[0] * i * (i - 1)/2. for i in x], clip_on=False, label='Snowball')
plt.plot(x, [b[0]* (i - 1) for i in x], clip_on=False, label='Linear')
plt.ylabel('Simple DMIs')
plt.xlabel('Substitution')
plt.legend(loc='upper left');
```

We compared the level of support for each model using Akaike's Information Criterion (AIC). If the difference in the AIC values ($\Delta$AIC) was greater than a threshold, we concluded that there was stronger support for the model with the lower AIC. 

```python
%%R
AIC(lin)
```

```python
%%R
AIC(orr)
```

This simulation provides stronger support for the linear model than for the snowball model ($\Delta$AIC = 13.9).
